```python
import numpy as np
from PIL import Image
img = np.array(Image.open('sample_two.png'))
half_width = img.shape[1] // 2
left_half = img[:, :half_width]
right_half = img[:, half_width:]
Image.fromarray(left_half).show()
Image.fromarray(right_half).show()
```

```python
img = np.array(Image.open('sample_two.png'))
flipped_img = np.fliplr(img)
Image.fromarray(flipped_img).show()
```

```python
import matplotlib.pyplot as plt
img = np.array(Image.open('sample_two.png'))
green_img = np.zeros_like(img)
green_img[:,:,1] = img[:,:,1]
plt.imshow(green_img)
plt.show()
```

```python
img = np.array(Image.open('sample_two.png'))
new_height = int(img.shape[0] * 0.8)
new_width = int(img.shape[1] * 0.8)
margin_height = (img.shape[0] - new_height) // 2
margin_width = (img.shape[1] - new_width) // 2
cropped_img = img[margin_height:margin_height+new_height, margin_width:margin_width+new_width]
Image.fromarray(cropped_img).show()
```
